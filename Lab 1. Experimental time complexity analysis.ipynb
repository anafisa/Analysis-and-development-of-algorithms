```python
import numpy as np
import time 
import random 
import matplotlib.pyplot as plt
import seaborn as sns  
sns.set_palette('Set2')
sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
def f_const(lst):
    return 1

def f_sum(lst):
    s = 0
    for i in lst:
        s += i
    return s

def f_product(lst):
    p = 0
    for i in lst:
        p *= i
    return p

def f_polynom(lst, x, n):
    s = 0
    for i in range(1, n+1):
        s += lst[i-1]*(x**(i-1))
    return s

def f_polynom_gorner(lst, x, n):
    s = 0
    for i in range(1, n+1):
        s += lst[i-1]*(x**(i-1))
    return s



def bubble_sort(lst):
    for i in range(len(lst)):
        for j in range(len(lst) - 1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst
    
    


# def quick_sort():


    
# def tim_sort():
        
        
    
    
```

```python
def calculate_time(f, v):
    start_time = time.time()
    f(v)
    return round(time.time() - start_time, 6)


def calculate_time1(f, v, x, n):
    start_time = time.time()
    f(v, x, n)
    return round(time.time() - start_time, 6)
```

```python
k = 500

result_const = []
result_sum = []
result_product = []
result_polynom = []
result_bubble = []
result_quick = []
result_tim = []

for n in range(1, k+1):
    res1 = []
    res2 = []
    res3 = []
    res4 = []
    res6 = []
    lst = np.random.randint(1, 100, n)
    for i in range(5):
        res1.append(calculate_time(f_const, lst))  
        res2.append(calculate_time(f_sum, lst))
        res3.append(calculate_time(f_product, lst))
        res4.append(calculate_time1(f_polynom, lst, 1.5, n))
#         res5 горнер
        res6.append(calculate_time(bubble_sort, lst))


        
    result_const.append(np.mean(res1))
    result_sum.append(np.mean(res2))
    result_product.append(np.mean(res3))
    result_polynom.append(np.mean(res4))
    result_bubble.append(np.mean(res6))


    
```
